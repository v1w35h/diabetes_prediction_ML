```python
train_df = pd.read_csv('/data/train.csv')
target = 'diagnosed_diabetes'

y = train_df[target]
X = train_df.copy()

X.drop(target,axis=1,inplace=True)

# print(y.head())
X.head()

```

# Splitting Data & One Hot encoding of categorical values

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
rng = 325

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=rng, stratify=y)


X_train_ohe = pd.get_dummies(X_train['gender'], dtype=int)
X_train_ohe = pd.concat([X_train_ohe, pd.get_dummies(X_train['ethnicity'], dtype=int)],axis=1)
# X_train_ohe = pd.concat([X_train_ohe, pd.get_dummies(X_train['education_level'], dtype=int)],axis=1)
# X_train_ohe = pd.concat([X_train_ohe, pd.get_dummies(X_train['income_level'], dtype=int)],axis=1)
X_train_ohe = pd.concat([X_train_ohe, pd.get_dummies(X_train['smoking_status'], dtype=int)],axis=1)
# X_train_ohe = pd.concat([X_train_ohe, pd.get_dummies(X_train['employment_status'], dtype=int)],axis=1)
# X_train_ohe.drop(['Other','Asian','No formal','High','Never','Unemployed'], axis=1, inplace=True)
X_train_ohe.drop(['Other','Asian','Never'], axis=1, inplace=True)

# X_train_ohe.head()

X_train.drop(['gender','ethnicity','education_level','income_level','smoking_status','employment_status'],axis=1,inplace=True)
# X_train.drop(['gender','ethnicity','education_level','income_level'],axis=1,inplace=True)
X_train = pd.concat([X_train, X_train_ohe],axis=1)
X_train.head()

X_val_ohe = pd.get_dummies(X_val['gender'], dtype=int)
X_val_ohe = pd.concat([X_val_ohe, pd.get_dummies(X_val['ethnicity'], dtype=int)],axis=1)
# X_val_ohe = pd.concat([X_val_ohe, pd.get_dummies(X_val['education_level'], dtype=int)],axis=1)
# X_val_ohe = pd.concat([X_val_ohe, pd.get_dummies(X_val['income_level'], dtype=int)],axis=1)
X_val_ohe = pd.concat([X_val_ohe, pd.get_dummies(X_val['smoking_status'], dtype=int)],axis=1)
# X_val_ohe = pd.concat([X_val_ohe, pd.get_dummies(X_val['employment_status'], dtype=int)],axis=1)
# X_val_ohe.drop(['Other','Asian','No formal','High','Never','Unemployed'], axis=1, inplace=True)
X_val_ohe.drop(['Other','Asian','Never'], axis=1, inplace=True)
# X_train_ohe.head()

X_val.drop(['gender','ethnicity','education_level','income_level','smoking_status','employment_status'],axis=1,inplace=True)
# X_val.drop(['gender','ethnicity','education_level','income_level'],axis=1,inplace=True)
X_val = pd.concat([X_val, X_val_ohe],axis=1)

X_train.head()
```

# Scaling data

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

mmscaler = MinMaxScaler(feature_range=(0, 1))

X_train['alcohol_consumption_per_week'] = mmscaler.fit_transform(X_train[['alcohol_consumption_per_week']])
X_val['alcohol_consumption_per_week'] = mmscaler.fit_transform(X_val[['alcohol_consumption_per_week']])

sScaler = StandardScaler()
ss = ['age',
      'physical_activity_minutes_per_week',
      'diet_score','sleep_hours_per_day',
      'screen_time_hours_per_day','bmi',
      'waist_to_hip_ratio','systolic_bp',
      'diastolic_bp', 'heart_rate',
      'cholesterol_total',
      'hdl_cholesterol', 'ldl_cholesterol',
      'triglycerides'
     ]

X_train[ss] = sScaler.fit_transform(X_train[ss])
X_val[ss] = sScaler.fit_transform(X_val[ss])

X_train.drop(['id'],axis=1,inplace=True)
X_val.drop(['id'],axis=1,inplace=True)

X_train.head()
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

cv_params = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga'], 
    'max_iter': [100, 500, 1000, 2000]
}

logreg_classifier = LogisticRegression(penalty = 'l2', C = 1, random_state = rng, max_iter = 1000)

log_cv = GridSearchCV(
    estimator = logreg_classifier,
    param_grid = cv_params,
    scoring='accuracy',
    cv=5,         
    refit='accuracy'     
)

log_cv.fit(X_train,y_train)
print(f"Optimal parameters: {log_cv.best_params_}")

# Predict on training set
y_hat = log_cv.predict(X_train)
print(f"Logistic Regression Classifier Model training accuracy is {accuracy_score(y_train, y_hat)*100:0.2f}%")
# Predict on validation set
y_hat = log_cv.predict(X_val)
print(f"Logistic Regression Classifier Model validation accuracy is {accuracy_score(y_val, y_hat)*100:0.2f}%")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

forest_size = [10, 50, 100, 200, 400, 800]
train_accuracies = []
val_accuracies = []

rf_model = RandomForestClassifier(n_estimators=10, bootstrap=True, min_samples_split=2, max_depth=50, min_samples_leaf=10)
rf_model.fit(X_train,y_train)

y_hat = rf_model.predict(X_train)
print(f"10 Random Forest Classifier Model training accuracy is {accuracy_score(y_train,y_hat)*100:0.2f}")
train_accuracies.append(accuracy_score(y_train,y_hat)*100)
y_hat = rf_model.predict(X_val)
print(f"10 Random Forest Classifier Model validation accuracy is {accuracy_score(y_val,y_hat)*100:0.2f}\n")
val_accuracies.append(accuracy_score(y_val,y_hat)*100)



rf_model = RandomForestClassifier(n_estimators=50, bootstrap=True, min_samples_split=2, max_depth=50, min_samples_leaf=10)
rf_model.fit(X_train,y_train)

y_hat = rf_model.predict(X_train)
print(f"50 Random Forest Classifier Model training accuracy is {accuracy_score(y_train,y_hat)*100:0.2f}")
train_accuracies.append(accuracy_score(y_train,y_hat)*100)
y_hat = rf_model.predict(X_val)
print(f"50 Random Forest Classifier Model validation accuracy is {accuracy_score(y_val,y_hat)*100:0.2f}\n")
val_accuracies.append(accuracy_score(y_val,y_hat)*100)



rf_model = RandomForestClassifier(n_estimators=100, bootstrap=True, min_samples_split=2, max_depth=50, min_samples_leaf=10)
rf_model.fit(X_train,y_train)

y_hat = rf_model.predict(X_train)
print(f"100 Random Forest Classifier Model training accuracy is {accuracy_score(y_train,y_hat)*100:0.2f}")
train_accuracies.append(accuracy_score(y_train,y_hat)*100)
y_hat = rf_model.predict(X_val)
print(f"100 Random Forest Classifier Model validation accuracy is {accuracy_score(y_val,y_hat)*100:0.2f}\n")
val_accuracies.append(accuracy_score(y_val,y_hat)*100)
```

```python
rf_model = RandomForestClassifier(n_estimators=200, bootstrap=True, min_samples_split=2, max_depth=50, min_samples_leaf=10)
rf_model.fit(X_train,y_train)

y_hat = rf_model.predict(X_train)
print(f"200 Random Forest Classifier Model training accuracy is {accuracy_score(y_train,y_hat)*100:0.2f}")
train_accuracies.append(accuracy_score(y_train,y_hat)*100)
y_hat = rf_model.predict(X_val)
print(f"200 Random Forest Classifier Model validation accuracy is {accuracy_score(y_val,y_hat)*100:0.2f}\n")
val_accuracies.append(accuracy_score(y_val,y_hat)*100)



rf_model = RandomForestClassifier(n_estimators=400, bootstrap=True, min_samples_split=2, max_depth=50, min_samples_leaf=10)
rf_model.fit(X_train,y_train)

y_hat = rf_model.predict(X_train)
print(f"400 Random Forest Classifier Model training accuracy is {accuracy_score(y_train,y_hat)*100:0.2f}")
train_accuracies.append(accuracy_score(y_train,y_hat)*100)
y_hat = rf_model.predict(X_val)
print(f"400 Random Forest Classifier Model validation accuracy is {accuracy_score(y_val,y_hat)*100:0.2f}\n")
val_accuracies.append(accuracy_score(y_val,y_hat)*100)



rf_model = RandomForestClassifier(n_estimators=800, bootstrap=True, min_samples_split=2, max_depth=50, min_samples_leaf=10)
rf_model.fit(X_train,y_train)

y_hat = rf_model.predict(X_train)
print(f"800 Random Forest Classifier Model training accuracy is {accuracy_score(y_train,y_hat)*100:0.2f}")
train_accuracies.append(accuracy_score(y_train,y_hat)*100)
y_hat = rf_model.predict(X_val)
print(f"800 Random Forest Classifier Model validation accuracy is {accuracy_score(y_val,y_hat)*100:0.2f}\n")
val_accuracies.append(accuracy_score(y_val,y_hat)*100)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(forest_size, train_accuracies,label='Training Accuracy')
plt.plot(forest_size, val_accuracies, label='Validation Accuracy')
plt.xlabel('Forest Size')
plt.ylabel('Accuracy')
plt.title('Random Forest: Accuracy vs Number of Trees')
plt.legend()
plt.grid(True)
plt.show()
```
